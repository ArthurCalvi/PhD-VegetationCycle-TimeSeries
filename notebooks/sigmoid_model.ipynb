```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, k, x0):
    return 1 / (1 + np.exp(-k * (x - x0)))

def vegetation_model(t, A, B, k1, x01, k2, x02):
    return A * (1 - sigmoid(t, k1, x01) + sigmoid(t, k2, x02)) + B

# Generate data points
t = np.linspace(0, 1, 1000)
A = 0.5   # Amplitude
B = 0.2   # Offset
k1 = 20   # Steepness of decrease
x01 = 0.3 # Midpoint of decrease
k2 = 20   # Steepness of increase
x02 = 0.7 # Midpoint of increase

y = vegetation_model(t, A, B, k1, x01, k2, x02)
s1 = sigmoid(t, k1, x01)
s2 = sigmoid(t, k2, x02)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(t, y, label='Vegetation Cycle')
plt.plot(t, s1, '--', label='Decrease Sigmoid')
plt.plot(t, s2, '--', label='Increase Sigmoid')
plt.xlabel('Time (normalized)')
plt.ylabel('Vegetation Index')
plt.title('Vegetation Cycle Model with Sigmoid Functions')
plt.legend()
plt.grid(True)
plt.show()
```
