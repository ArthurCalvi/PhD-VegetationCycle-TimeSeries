```python
import numpy as np
import matplotlib.pyplot as plt

def sinusoidal_model(t, A1, A2, A3, phi1, phi2, phi3):
    omega = 2 * np.pi  # Angular frequency for a period of 1
    return (A1 * np.sin(omega * t + phi1) + 
            A2 * np.sin(omega * t + phi2) + 
            A3 * np.sin(omega * t + phi3))

# Generate time points for one cycle
t = np.linspace(0, 1, 1000)

# Parameters for the three sinusoids
A1, A2 = 0.5, 0.5  # Amplitudes for the two main sinusoids
A3 = -0.3  # Amplitude for the counteracting sinusoid
phi1, phi2 = 0, np.pi / 4    # Phases for the two main sinusoids
phi3 = (phi1 + phi2) / 2   # Phase for the counteracting sinusoid

# Calculate individual components and the combined signal
y1 = A1 * np.sin(2 * np.pi * t + phi1)
y2 = A2 * np.sin(2 * np.pi * t + phi2)
y3 = A3 * np.sin(2 * np.pi * t + phi3)
y_combined = sinusoidal_model(t, A1, A2, A3, phi1, phi2, phi3)

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, y1, label='Sinusoid 1', alpha=0.5)
plt.plot(t, y2, label='Sinusoid 2', alpha=0.5)
plt.plot(t, y3, label='Counteracting Sinusoid', alpha=0.5)
plt.plot(t, y_combined, label='Combined Signal', linewidth=2)
plt.plot(t, y1 + y2, label='Sum of Sinusoids 1 & 2', linestyle='--', alpha=0.5)

plt.title('Approximation of Piecewise Linear Function Using Sinusoids')
plt.xlabel('Time (normalized)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
